These vaery similar plots allow you to aggregate data off a categorical feature in the data. *barplot* is a general plot that allows you to aggregate categorical data based off some function, by default the 'mean'.

# Box Plot - Outlier Representation
Quartile measurements OR Outlier measurements: boxplot and violin plots are used to show the distribution of categorical data. A box plot (or a box and whisker plot) shows the distribution of quantitive data in a way that facilitates comparisons between variables or across levels of categorical variables. The box shows the quarlites of the data set while the whiskers extend to show the rest of the distribution, except for the points that are determined to be "outliers" using a method that is a function of the inter-quartile range. 
- Outliers are those unusual values which can be real values as well as some spurious data points.
- As long as we are doing data analysis/insghts, they are fine. But when we build ML models, it is important to spell them out.

```python
'''
Q1 - 25 PERCENTILE 
Q3 - 75 PERCENTILE
IQR = Q3 - Q2
MAX_UPP = Q3 + 1.5 * IQR
MIN_LOW = Q1 - 1.5 * IQR

np.percentile(tips['total_bill'], [25, 75])
'''
```

```python
q1, q3, med = np.percentile(tips['total_bill'], [25, 75, 50])
print("Lower quartile {}".format(q1))
print("Median quartile {}".format(med))
print("Upper quartile {}".format(q3))

iqr = q3 - q1
print("Inter quartile range {}".format(iqr))

upp = iqr * 1.5 + q3
low = q1 - iqr * 1.5

print("Upper limit {}".format(upp))
print("Lower limit {}".format(low))

sns.boxplot(data = tips['total_bill'])
plt.show()

sns.boxplot(data = tips['tip'])
plt.show()
```
