# Machine Learning (ML)
Based on the historical data making the prediction of the future is one of the domains of machine learning (ML). 

Last 10 years of data of all employees who have left and who are still working. 
- L employees left - # of promotions, increments, working hrs, wfh, wlb, leaves, tenure, etc. 
- P employees present - # of promotions, increments, working hrs, wfh, wlb, leaves, tenure, etc. 
- N new hires

- Current year (2023): P + N
    - Find out the pattern from the historical data to find out the patterns from the L and P data set for emplpoyees either leaving or still working in the co. This process is called as exploratory data analysis (EDA).
    - Various tools such as pandas, matplotlib, numpy, etc. are used to perform the EDA.
    - Using the EDA, get the data insights, and based on it come up with the story line. 
    - EDA is always done on the past or historical data. 

## Process of building ML model
    1. Import the data
    2. Perform the EDA (find the most important factors, generate the data insights, understand the data set)
    3. List the key features / independent variables / (a.k.a. X) (e.g., increments, working hrs, wfh, etc.)
    4. List the key goals / dependent variables / (a.k.a. Y) (leave or not leave the company)
    5. Data preprocessing (bring the X into correct format)
        - Object to number transformation
        - Outlier and missing data processing
        - Feature engineering (creating more features)
        - Data cleaning (special chars, missing data. etc.)
    6. Schufle and split the data into "training data" and "test data".
         - Schuffle the data set to avoid any biases or patterns in the data.
         - 80, 75, 70 % of training data
            - Provide the x_train and y_train to the ML model and train it. 
        - 20, 25, 30 % of test data. 
            - Provide the input x_test to the model and predict the output value.
            - Compare the predicted value coming from the model with actual value already abvialble in the data (y_test). 
        - The difference between the two will help understand the correctness of the model. 
    7. Build the ML model to do predictions (as human may be biased, not attentive, prone to errors, etc.)
        - Train the ML model
        - Predict the future (Y) based on input features (X)
    8. Evaluate the ML model (correctness)
    9. Deploy the model (real-time processing)

    - The steps 2 to 7 are kind of iterative steps till the model does not give the right prediction. 
    10. Once the mode is defined with a given set of features, that model is never changed for subsequent (future) data set. If for whatever reason, new features are to be considered, a new model needs to be designed. 


- Questions:
    - Even though training is based on historical data, the current (ongoing) state of the features is still evolving and not good enough to do a more correct prediction of the Y. What is the value of such preductions? 
    - What is the threshold of X at which we can say the model gives more reasonalble / logical Y?

# Categories of ML
- Supervised ML: The set of features (X) and respective predictions (Y) are provided upfront. 
    - Classification models
        - prediction of a categorical value (Employee stays/leaves, product purchased/not-purchased/returned, load approved/rejected/review, etc.)
        - More the classes in the output, the less performant model 
    - Regression models
        - Prediction of a contineous value (sales amount, GDP, interest rate, loan amount, etc.)
- Unsupervised ML: only the input features (X) are provided and no Y is known. 

# Linear Regression Algorithm
Linear regression is a machine learning algorithm based on supervised learning. It performs a regression task. 

Linear regression performs the task to predict a dependent variable value (y) based on a give independent variable (x)

It is mostly used for finding out the relationship between variables and forecasting. 

- Intercept = constant value - base value of y when no features contribute i.e., y whe x = 0
- Coefficient / slopes / beta values: change in y when 1 unit value change in x. 
- predicted y = intercept + coefficient 1 * feature 1 + coefficient 2 * feature 2 + ... + coefficient n * feature n

## Training the model
- The training phase may take longer time depending on the size of data and the number of input features. 

```python
# Import the linear regression algorithm
from sklearn.linear_model import LinearRegression

# Create a LinearRegression instance
lm = LinearRegression()

# Train the model with the input feature data
lm.fit(x_train, y_train)
```

## Evaluating the model

```python
# See the difference between test output and the predicted output
pd.DataFrame({'test value': y_test, 'predicted value': predictions})
```

### Mean square error (MSE)
<img src="https://www.gstatic.com/education/formulas2/472522532/en/mean_squared_error.svg" width="450"/>

- MSE value range from 0 to inf. 
- The value close to 0, better the model
- Very sensitive to outliers (because of the square of the difference between observed and predicted value)

### Mean absolute error (MAE)
<img src="https://www.gstatic.com/education/formulas2/472522532/en/mean_absolute_error.svg" width="450"/>

- MAE value range from 0 to inf. 
- The value close to 0, better the model
- It is not sensitive to outliers (because of the difference between observed and predicted value)

### Root mean square error (RMSE)
<img src="https://www.gstatic.com/education/formulas2/472522532/en/root_mean_square_deviation.svg" width="450"/>


### R2 square error
<img src="https://vitalflux.com/wp-content/uploads/2019/07/R-squared-formula-function-of-SSE-and-SST.jpg" width="400"/>

- MAE value range from 0 to 1. 
- The value close to 1, better the model


```python
# Import the linear regression algorithm
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print('Mean absolute error: ', mean_absolute_error(y_test, predictions))

print('Mean squared error: ', mean_squared_error(y_test, predictions))

print('Root mean absolute error: ', np.sqrt(mean_squared_error(y_test, predictions)))

print('R Square error: ', r2_score(y_test, predictions) * 100)
```

- Generally, the model is not so good. However, in the context of the scale of the test (y_test) values, the difference is going to be huge. Therefore the mean is going to be huge. Given the range of the y_test, the difference is justified.

### Fit of the model
- Like we calculated score for testing data, we need to calculate the score for trainign data. To achieve this, we need to predict based on the training data.
- We use training data to train the model followed by test data to test the effectiveness. The effectiveness is the error (or the score) that model work with. 
- Now we have to find out if the model has crammed the test data (input syllabus). For this we have to judge the model using the training data and find out the error (or the score). 

```python
train_predictions = lm.predict(x_train)
train_predictions = np.round(train_predictions, 2)
train_predictions

print('Train Mean absolute error: ', mean_absolute_error(y_train, train_predictions))

print('Train Mean squared error: ', mean_squared_error(y_train, train_predictions))

print('Train Root mean absolute error: ', np.sqrt(mean_squared_error(y_train, train_predictions)))

print('Train R Square error: ', r2_score(y_train, train_predictions) * 100)
```

- What we observing that model performs well (i.e., less error) using the data which was used to train the model. However, this is a model whose behavior changes for different data inputs. 
- Ideally, we need a model which behaves consistently for variety of data provided to it. 

- The concept of checking the model performance is called as **overfillting** and **underfitting** model. 

### Overfitting model (High variance)
- Noticiable difference between the training score and the test score, but they are on higher side. For example:
    - training score (95) & test score (90): (not accepted) as difference is too big (> 2).
    
### Underfitting model (High bias)
- None of the test data score and training data score are not good (rather very low).
    - training score (30) & test score (35): not accepted as scores are very low.
    
### Good fitting model
- training score ~ test score: accepted and model is good fitting. (The difference is ~ 1)
- training score (95) & test score (94)
- training score (95) & test score (94.5)
- training score (95) & test score (95)

#### Notes:
- High variance
    - Not causing any error to happen. The regression line follows each and every value given. 
- High bias
    - Model is giving some random prediction. There are too many errors in the model. 
- Ideal model (good fitting model)
    - Not too perfect that it captures every data point and not too loose that it gives all random scores. 
- Whether the model is good or bad depends on the business use case and what is the apetite for the error.
